# An initial Graph SBM test

```julia
import Pkg
Pkg.add("Conda")
using Conda
Conda.add("networkx")
Pkg.add("PyCall")
using PyCall
nx = pyimport("networkx")
Pkg.add("PyPlot")
using PyPlot
```

### Generate a small SBM with *very* distinct communities

```julia
sizes = [10, 10, 20]
probs = [[0.7, 0.05, 0.02], [0.05, 0.7, 0.07], [0.02, 0.07, 0.7]]
g = nx.stochastic_block_model(sizes, probs, seed=0)
```

```julia
A = nx.to_numpy_matrix(g)
imshow(A)
```

```julia
using LinearAlgebra
F = svd(A)
```

```julia
imshow(F.U)
colorbar()
```

```julia
imshow(F.Vt)
colorbar()
```
